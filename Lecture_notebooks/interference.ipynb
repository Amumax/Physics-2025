# Интерференция

## 1. Понятие интерференции

Когда электромагнитная волна падает какую-либо регистрирующую поверхность (фотопластинку, фотоматрицу, сетчатку глаза, фотоприёмник и т.п.), то на ней регистриуется не *мгновенное* значение напряжённости волны, а *средняя плотность потока энергии* (*интенсивность*) волны. 

**Замечание.* Напомним, что среднее значение функции $f(t)$ за промежуток времени $\Delta t$, по определению, равно
$$
\left< f \right>(t) = \int\limits_{t}^{t+\Delta t} f(t) dt.
$$
Как правило, время усреднения на много порядков превосходит переиод колебания волны $\Delta t \lll T$. Например, для человеческого глаза $\Delta t\sim 1/25$ с, тогда как период волны для видимого света (средняя длина волны $\lambda\sim 500$ нм составляет $T \sim \frac{500\cdot 10^{-9}\;\text{нм}}{3 \cdot 10^8 \text{м/с}} \sim 10^{-15}$ с.

Оговоримся также, что среднее значение произвольной гармонической функции, период которой значительно превышает время усреднения, равно *нулю*:
$$
\left<\cos(\omega t + \varphi)\right> = 0,\qquad \text{при } \omega \gg 2\pi /\Delta t.
$$

Как известно, плотность потока энергии световой волны пропорциональна квадрату напряжённости электрического поля: $I \propto \left<E^2\right>$. В *данном разделе* для краткости примем обозначение $$
I=\left<E^2\right>,
$$ 
то есть будем называть "интенсивностью" просто среднее значение квадрата напряжённости поля.

Рассмотрим cумму двух произвольных волн с напряжённостями $\vec{E}_1$ и $\vec{E}_2$. Напряжённости волн, ввиду линейности уравнений электромагнитного поля (*принцип суперпозиции*), складываются:
$\vec{E} = \vec{E}_1 + \vec{E}_2$. Тогда "интенсивность" суммарной волны равна
$$
I = \left<|\vec{E}_1 + \vec{E}_2|^2 \right> = I_1 + I_2 + 2\left<(\vec{E}_1 \vec{E}_2)\right>.
$$
Если последнее слагаемое
$$
I_{int} = 2\left<(\vec{E}_1 \vec{E}_2)\right>
$$
не обращается в нуль, то говорят, что имеет место явление **интерференции** волн.

Заметим, что если расссматривать величины $E_1$ и $E_2$ как некоторые общие случайные функции с нулевым средним ($\left<E_1\right>=\left<E_2\right>=0$), то с точки зрения теории вероятностей
интерференционное слагаемое $I_{int}$ представляет не что иное, как *корреляционную функцию* двух величин. В оптике вместо термина *корреляция* используется термин *когерентность*, который означает *сохранение разности фаз между двумя колебательными процессами*.
## 2. Условия интерференции

Сформулируем необходимые условия, при которых интерференционное слагаемое будет ненулевым.

1. Интерференция возможно только между компонентами волны с *одинаковой поляризацией*.

Действительно, если поляризации ортогональны, $\vec{E}_1\perp \vec{E}_2$, то их скалярное произведение тождественно равно нулю: 
$(\vec{E}_1 \vec{E}_2)=0$.

2. Для гармонических волн интерференция возможна только если частоты волн *одинаковы*.

Действительно, если колебания будут иметь разные частоты $\omega_1$ и $\omega_2$, то при вычислении 
среднего значения произведения функций $\left<(\vec{E}_1 \vec{E}_2)\right>$
возникнут множители вида 
$$
\cos \omega_1 t \cdot \cos \omega_2 t = \frac{1}{2} \cos((\omega_1+\omega_2)t) + \frac{1}{2} \cos((\omega_1-\omega_2)t) .
$$
Однако, как замечено выше, среднее значение любой гармонической функции следует считать равным нулю: $\left<\cos \omega t\right> = 0$.

Таким образом, интерферецния может возникнуть только если частоты волн совпадают $\omega_1 = \omega_2$ (точнее, если разность частот не превысит частоту усреднения: $|\omega_1 - \omega_2| < 2\pi / \Delta t$).

### 2.1. Интерференция монохроматических колебаний

Рассмотрим монохроматические колебания полей от двух волн в некоторой фиксированной точке пространства. Их можно представить как
$$
\vec{E}_1 = \vec{A}_1 \cos(\omega t + \varphi_1),\quad \vec{E}_2 = \vec{A}_2 \cos(\omega t + \varphi_2),
$$
где $\varphi_1$  и $\varphi_2$ --- фазы волн в данной точке (зависящие, конечно, от положения точки).

Вычислим интерференционное слагаемое (ср. с формулой для *мощности цепи переменного тока*):
$$
I_{int} = 2(\vec{A}_1\vec{A}_2) \left<\cos(\omega t + \varphi_1)  \cos(\omega t + \varphi_2)\right>= \boxed{(\vec{A}_1\vec{A}_2) \cos \Delta \varphi}.
$$
(здесь мы использовали формулу произведения косинусов и учли, что $\left<\cos (2\omega t+\varphi_1+\varphi_2) \right>=0$).

Отметим также эту же формулу в комплексном представлении (рекомендуем проверить самостоятельно):
$$
I_{int} = \mathrm{Re} (\vec{A}_1 \vec{A}_2^{\star}).
$$

Таким образом, задача об интерференции сводится к задаче о вычислении *разности фаз* волн.

Если волна имеет несколько спектральных компонент (несколько частот $\omega_i$), то каждая компонента даст *независимый* вклад в интерференционное слагаемое (волны разны частот не интерферирурют!).




### 2.2. Максимумы и минимумы интерференции

Интенсивность достикает максимума при $\cos \Delta \varphi = 1$, то есть при $\Delta \varphi = 2\pi m$, $m\in \mathbb{Z}$. Она равна
$$
I_{max} = (\vec{A}_1 + \vec{A}_2)^2
$$

Минимумы: $\cos \Delta \varphi = -1$, $\Delta \varphi = \pi (2 m + 1)$, $m\in \mathbb{Z}$.
$$
I_{min} = (\vec{A}_1 - \vec{A}_2)^2.
$$

Значение $m = \Delta \varphi / (2\pi)$ называют *порядком интерференции*.


### 2.3. Видность (контрастность) картины

*Видностью* (или *контрастностью*) интерференции называют отношение
$$
V = \frac{I_{max}-I_{min}}{I_{max}+I_{min}}.
$$

Наилучшая видность $V=1$ достигается при $I_{min}=0$, что возможно, если волны имеют одинаковую амплитуду $A_1 = A_2$ (то есть волны могут в точности скомпенсироваться).

Если амплитуды сильно различаются ($A_1 \ll A_2$ или $A_1 \gg A_2$), то $I_{max} \approx I_{min}$ и видность интерференции стремится к нулю $V\to 0$.


## 3. Интереференция монохроматических волн

### 3.1. Интерференция плоских волн. Ширина интерференционной полосы
Рассмотрим две плоские монохроматические волны с одинаковыми частотами $\omega$, но разными по направлению волновыми векторами $\vec{k}_1$ и $\vec{k}_2$
(по модулю одни одинаковы: $|\vec{k}_1|=|\vec{k}_2| = \frac{2\pi }{\lambda} = \frac{\omega}{v}$, где $\lambda$ --- длина волны в среде, $v$ --- скорость волны в среде). Амплидуты каждой волны примем для простоты за единицу: $|A_1|=|A_2|=1$.

Фазы этих волн имеют вид: $\varphi_1 = (\vec{k}_1 \vec{r})$, $\varphi_2 = (\vec{k}_2 \vec{r})$ (для простоты считаем, что фазы волн совпадают в начале координат). 

Разность фаз имеет вид:
$$
\Delta \varphi = (\vec{K} \cdot \vec{r}),
$$
где $\vec{K} = \vec{k}_2 - \vec{k}_1$ --- "разностный" волновой вектор.

Разместим экран перпендикулярно суммарному вектору $\vec{k}_1+\vec{k}_2$. Направим ось $x$ вдоль вектора $\vec{K}$. 
Интерференционное слагаемое примет вид
$$
I_{int} = 2 \cos (K x),
$$
где 
$$
K = |\vec{k}_1-\vec{k}_2| = \frac{2\pi\sin \frac{\alpha}{2}}{\lambda},
$$
$\alpha$ --- угол между векторами $\vec{k}_1$ и $\vec{k}_2$ (угол схождения волн).

На экране возникнут эквидистантные *интерференционные полосы*. Расстояние между максимумами полос ("ширина" полос) равно:
$$
\Delta x = \frac{2\pi}{K} = \frac{\lambda}{2\sin \frac{\alpha}{2}}.
$$
Если угол схождения мал, $\alpha \ll 1$ то
$$
\Delta x \approx \frac{\lambda}{\alpha}.
$$

```python
N=1000
Lambda = 1
alpha = 1e-2
K = 2 * np.pi * np.sin(alpha/2) / Lambda
XMAX = 5 * 2 * np.pi / K
x = np.linspace(-XMAX,XMAX,N)
Iint = np.cos(K * x)

A1 = 1
A2 = 1
Imax = (A1 + A2)**2
I = A1**2 + A2**2 + 2 * A1 * A2 * Iint
plt.imshow(I * np.ones(100)[:, np.newaxis], vmin=0, vmax=Imax, cmap='gray')
plt.title("Интерференционные полосы, видность $V=1$")
plt.show()
```

Если амлидуты волн различаются, то полной компенсации при интерференции не будет и картина получится менее контрастной.

```python
A1 = 1
A2 = 0.25
Imax = (A1 + A2)**2
Imin = (A1 - A2)**2
I = A1**2 + A2**2 + 2 * A1 * A2 * Iint
plt.imshow(I * np.ones(100)[:, np.newaxis], vmin=0, vmax=Imax, cmap='gray')
plt.title(f"Интерференционные полосы, видность $V={(Imax-Imin)/(Imax+Imin):.2f}$")
plt.show()
```

### 3.2. Оптическая разность хода

Предположим, что для распространения света применимо приближение геометрической оптики.
То есть путь любого участка фронта волны можно проследить вдоль некоторого непрерывного (возможно, искривлённого) луча.
Разность фаз, которую наберёт волна вдоль такого луча будет равна
$$
\Delta \varphi = \int (\vec{k} d\vec{l}) 
$$
где интегрирование ведётся по отрезкам луча $d\vec{l}$. 

Рассмотрим вличину $\ell = \frac{\Delta \varphi}{k_0}$, где $k_0=c/\omega_0$ --- волновое число в пустое. С учётом того, что в среде с показателем преломления $n$ волновой вектор увеличивается в $n$ раз (длина волны уменьшается в $n$ раз), а луч всегда направлен вдоль вектора $\vec{k}$, получим:
$$
\ell' = \int n\, dl,
$$
где интегрирование ведётся вдоль луча. Эту величину называют *оптической длиной пути*. Разность двух оптических длин называют *оптической разностью хода*.

Разность фаз лучей, прошедших по два разных пути, кратна отношению оптической разности хода к длине волны в пустоте:
$$
\Delta \varphi = 2\pi \frac{\ell'}{\lambda_0}.
$$

**Потеря полуволны при отражении**. Если волна отражается от оптически более плотной среды, то, согласно формулам Френеля, происходит изменение фазы отражённой волны на $\Delta \varphi=\pi$.
В таком случае при расчёте оптической длины к ней необходимо добавить (или вычесть, что эквивалентно) половину длины волны $\ell_{отр} = \lambda/2$. При отражении от оптически менее плотной среды фаза не меняется и такую добавку длеать не нужно.

### 3.3. Интерференция от двух точечных источников

Рассмотрим два точечных монохроматических источника, расположенных на расстоянии $d$ друг от друга. 

Экран наблюдения расположен в плоскости $z=L$, $L\gg d$.

#### Случай А: интерференция плоских волн

Разместим источники вдоль линии, параллельной экрану: координаты источников ($0,-d/2,0$) и ($0,+d/2,0$).

Разность хода лучей равна:
$$
\ell' = \sqrt{x^2 + (y-d/2)^2 + L^2} - \sqrt{x^2 + (y+d)^2 + L^2}.
$$
Пусть $d\ll \sqrt{x^2+y^2} \ll L$ (точки вблизи центра экрана). Тогда после разложения по Тэйлору:
$$
\ell' \approx \frac{(x-d/2)^2}{2L} - \frac{(x+d/2)^2}{2L} \approx \frac{xd}{L}.
$$
Видно, что в центре экрана наблюдается описанная выше *интерференция двух плоских волн*, сходящихся под углом $\alpha \approx d / L$. Ширина полос там $\Delta x = \lambda L / d = \lambda/\alpha$.

```python
N=1000
Lambda = 5e-4  # длина волны, мм
k = 2 * np.pi / Lambda  # волновое число
d = 8e-3  # расстояние между источниками, мм
L = 10    # расстояние до экрана, мм
W = 20    # размеры наблюдаемой области на экране
H = 20

X0 = np.linspace(-W,W,N)
Y0 = np.linspace(-H,H,N)

X, Y = np.meshgrid(X0, Y0)

D = np.sqrt(X**2 + (Y - d/2)**2 + L**2) - np.sqrt(X**2 + (Y + d/2)**2 + L**2)
Iint = 2 * (1 + np.cos(k * D))
plt.imshow(Iint, vmin = 0, vmax = 4, cmap='gray')
plt.title("Интерференция от двух точечных источников -- схема Юнга")
plt.show()
```

#### Случай Б: интерференция сферических волн

Разместим источники вдоль линии, перпендикулярной экрану в точки ($0,-d,0$) и ($0,0,0$).

Теперь 
$$
\ell' = \sqrt{r^2 + (L+d)^2} - \sqrt{r^2 + L^2},\qquad r = \sqrt{x^2+y^2}.
$$
Пусть $r\ll L$. Тогда после разложения по Тэйлору:
$$
\ell' \approx d \left(1-\frac{r^2}{2L(L+d)}\right).
$$
На экране будет наблюдаться система тёмных и светлых *колец* (аналог колец Ньютона). Предположим, что $d/\lambda=m_0$ --- целое число. Тогда в центре картины будет наблюдаться максимум, 
а остальные светлые полосы будут иметь радиусы
$$
r_m = r_1 \sqrt{m}, \qquad m = 1,2,\ldots
$$
($r_1 = \sqrt{2\lambda L\frac{L+d}{d}}$).

```python
N=1000

Lambda = 5e-4
k = 2 * np.pi / Lambda
d = 0.5 + 0*Lambda/2
L = 1
W = 0.1
H = 0.1

X0 = np.linspace(-W,W,N)
Y0 = np.linspace(-H,H,N)

X, Y = np.meshgrid(X0, Y0)

D = np.sqrt(X**2 + Y**2 + (L-d/2)**2) - np.sqrt(X**2 + Y**2 + (L+d/2)**2)
Iint = 2 * (1 + np.cos(k * D))
plt.imshow(Iint, vmin = 0, vmax = 4, cmap='gray')
m0 = int(d/Lambda)
for m in range(1,10):
    plt.gca().add_patch(plt.Circle((N/2,N/2), np.sqrt(m) * np.sqrt(2*Lambda*(L-d/2)*(L+d/2)/d) * N/W/2, color='red', fill = False,lw=0.5))
plt.title("Интерференция от двух точечных источников -- ''кольца Ньютона''")
plt.show()
print("Красные линии -- теоретические положения максимумов.")
```

**Замечание**: упомянутые *кольца Ньютона* соответствуют интерференции *плоской* и *сферической* волны. Это соответствует пределу $d\to \infty$: один из источников удаляется на бесконечность и сферическая волна заменяется на плоскую. Тогда при $d\to \infty$ получим
$$
\ell' \approx d -\frac{r^2}{2L}.
$$
Здесь первое слагаемое формально бесконечно, поэтому его можно считать кратным $\lambda$ и, следовательно, просто отбросить. Радиусы (светлых) колец Ньютона:
$$
r_m = \sqrt{2\lambda L m}, \qquad m = 1,2,\ldots
$$
где $L$ --- радиус кривизны сферической волны. Если сферическая волна рождается в результате отражения плоской волны от сферической поверхности радиуса $R$, то $L=R/2$ (фокус сферического зеркала) и, следовательно,
$$
r_m = \sqrt{\lambda R m}, \qquad m = 1,2,\ldots
$$

## Интерференция немонохроматического света

Пусть в спектре есть две частоты $\omega_1$ и $\omega_2=\omega_1 + \Delta \omega$, то они дадут *независимые* интерференционные картины:
$$
I_1 = 2I_0 (1 + \cos \frac{\omega_1}{c}\ell),\qquad I_1 = 2I_0 (1 + \cos \frac{\omega_2}{c}\ell)
$$

Максимумы одной картины наложатся на минимумы другой при
$$
\frac{\Delta \omega}{c} \ell = \pi.
$$

Выразим $\Delta \omega = \Delta (\frac{2\pi c}{\lambda}) \approx \frac{2\pi c}{\lambda^2}\Delta \lambda$ и учтём, что $m=\ell/\lambda$. Отсюда условие *пропадания* интерференционной картины:
$$
m = \frac{1}{2}\frac{\lambda}{\Delta \lambda}
$$
или 
$$
\ell = \frac{\lambda^2}{2\Delta \lambda}.
$$

В схеме Юнга $m$ --- это не только порядок интерференции, но и номер интерференционной полосы (поскольку для центральной полосы $m=0$). Поэтому полоса с таким номером практически не будет видна (будет иметь нулевую видность).

Если частоты действительно всего *две*, то видность будет периодически восстанавливаться (подобно *биениям* суммы двух косинусов). 

В общем случае спектр занимает некоторую сплошную полосу $\Delta \omega$. Для получения ответа нужно нужно разбить спектральный интервал на узкие полосы и проинтегрировать по ним. Эффекта восстановления видности в общем случае не будет (хотя он может проявляться для особых спектральных распределений, например для равномерного, как в примере ниже). Но полосы в любом случае пропадут при  
$$
m > m_{max} \sim \frac{\lambda}{\Delta \lambda}.
$$

```python
### Влияние немонохроматичности

N=1000
Lambda = 5e-4  # длина волны, мм
k = 2 * np.pi / Lambda  # волновое число
d = 10e-3  # расстояние между источниками, мм
L = 10    # расстояние до экрана, мм
W = 20    # размеры наблюдаемой области на экране
H = 20

NK = 100
D_Lambda = 0.1 * Lambda
D_k = D_Lambda / Lambda * k

X0 = np.linspace(-W,W,N)
Y0 = np.linspace(-H,H,N)

X, Y = np.meshgrid(X0, Y0)

D = np.sqrt(X**2 + (Y - d/2)**2 + L**2) - np.sqrt(X**2 + (Y + d/2)**2 + L**2)

PHI = k * D  

Iint = np.zeros_like(PHI)
for _k_ in np.linspace(k-D_k, k+D_k, NK):
    Iint += 2 * (1 + np.cos(_k_ * D))
Iint /= NK

plt.imshow(Iint, vmin = 0, vmax = 4, cmap='gray')
plt.title("Исчезновение интерференции в немонохроматическом свете")
plt.show()
```

### Когерентность, модель цугов

Излучение реального источника никогда не представляет собой чистую синусоиду. Каждый отдельный атом излучает излучает сигнал, очень близкий к идеальной синусоиде, однако он это делает в течение некоторого *конечного* времени. Таким образом, излучение большого количества атомов представляет собой **сумму** большого количества синусоид, начинающихся в случайные моменты времени 

Из соотношения неопределённостей для спектров делаем вывод, что излучение любого атома не строго монохроматическое, а имеет спектр с характерной шириной $\Delta \omega \sim 2\pi /\tau$ (или $\Delta \nu \sim 1/\tau$). За время излучения цуга атом успевает совершить большое количество колебаний ($N=\tau/T \gg 1$), поэтому спектр цугов узкий: $\Delta \omega \ll \omega$.

Условие существования интерференции можно переписать в следующем виде:
$$
\Delta \omega \ell / c < \pi\qquad \to \qquad \ell < \frac{1}{2} c \tau.
$$
Величина $c\tau=N\lambda$ --- это пространственная "длина" цуга. Интерференция возможна, если *один и тот же цуг* перекрывается *сам с собой* при наличии разности хода $\ell$. 

Этому есть простое объяснение: разные атомы излучают цуги со *случайной* начальной фазой. Следовательно, цуги от разных атомов в среднем не могут интерферировать ---среднее значение их интерференционного слагаемого $\left<\cos\Delta \varphi\right>$ будет равно нулю! Итак, каждый отдельный цуг может интерферировать только *сам с собой*.

Можно сказать, что суммарная *фаза* волны случайным образом сбивается с интервалом порядка $\tau$. 

Постоянство разности фаз любых двух волн называют **когерентностью**. 

Характерное время "забывания" фазы $\tau_{ког}$ называется *временем когерентности*.


```python
maxt = 10.0
T = 1e-2 * maxt
omega = 2 * np.pi / T
tau = 10 * T # длительность цуга
N = 10
numpoints=int(1e6)
t0 = np.random.uniform(0,maxt,N) # равномерное распределение начал цугов
t = np.linspace(0.0,maxt, numpoints)

def zug(t, t0, tau):
    return np.cos((t - t0) * omega) * (np.abs(t - t0) < tau)

s = np.zeros_like(t)
for _t0 in t0:
    z = zug(t, _t0, tau)
    plt.plot(t, z)
plt.title(f"Набор случайных цугов (пример $N={N}$)")
plt.xlabel("$t$")
plt.ylabel("$E(t)$")
plt.show()

N = 1000
t0 = np.random.uniform(0,maxt,N) # равномерное распределение начал цугов
s = np.zeros_like(t)
for _t0 in t0:
    z = zug(t, _t0, tau)
    s += z
plt.plot(t, s)
plt.xlabel("$t$")
plt.ylabel("$E(t)$")
plt.title(f"Сумма сучайных цугов ($N={N}$)")
plt.show()
```

```python
from scipy.fft import fft, fftfreq
C = fft(s)
plt.plot(fftfreq(numpoints, maxt)[:N//2], abs(C)[:N//2] / N)
plt.xlabel(f"Частота, $\\nu$")
plt.ylabel(f"Амплитуда гармоники, $a_{{\\nu}}$")
plt.title(f"Спектр суммы цугов ($N={N}$, $\\tau = {tau/T:.1f}T$)")
plt.show()
```

### Влияние конечного размера источника

Разные точки источника также излучаеют *некогерентно* друг другу. Следовательно, каждая точка источника даёт независимую картину интенсивности, которая складывается с остальными. Если от разность хода интерферрирующих лучей зависит от координаты точки на источнике, то интерференционные картины могут наложиться друг на друга и пропасть.

Простейший пример: если источник в схеме Юнга состоит ровно из двух некогерентных точек, расположенных на расстоянии $b$, то между ними возникнет дополнительная разность хода $\ell_0 \approx b \cdot \frac{d}{L_0}$. Если эта разность хода превысит $\lambda/2$, картина пропадёт. Таким образом, дополнительное условие интерференции:
$$
b < \frac{\lambda}{2} \cdot \frac{L_0}{d}
$$
Если определить угловой размер источника как
$\psi \sim \frac{b}{L_0}$, получаем 
$$
d < \frac{\lambda}{2\psi}.
$$
Таким образом, при конечном угловом размере источника $\psi$ щели в опыте Юнга должны находиться на достаточно малом расстоянии. 

Величину $\rho_{ког} = \lambda/\psi$ называют *радиусом пространственной когерентноти* --- он определяет размер "пятна", в пределах которого колебания поля можно считать когерентными друг другу.

```python
N=1000
Lambda = 5e-4  # длина волны, мм
k = 2 * np.pi / Lambda  # волновое число
d = 5e-3  # расстояние между источниками, мм
L = 10    # расстояние до экрана, мм
W = 20    # размеры наблюдаемой области на экране
H = 20

X0 = np.linspace(-W,W,N)
Y0 = np.linspace(-H,H,N)

X, Y = np.meshgrid(X0, Y0)

D = np.sqrt(X**2 + (Y - d/2)**2 + L**2) - np.sqrt(X**2 + (Y + d/2)**2 + L**2)

PHI = k * D  


B = 1.0  # размер источника
SL = 10   # расстояние до источника
SN = 50

P0 = np.linspace(-B/2,B/2,SN)
Q0 = np.linspace(-B/2,B/2,SN)

P, Q = np.meshgrid(P0, Q0)

D0 = np.sqrt(P**2 + (Q - d/2)**2 + SL**2) - np.sqrt(P**2 + (Q + d/2)**2 + SL**2) 

Iint = np.zeros_like(PHI)
for _D0_ in D0.flat:
    Iint += 2 * (1 + np.cos(PHI + k * _D0_)) 
Iint /= D0.size

plt.imshow(Iint, vmin = 0, vmax = 4, cmap='gray')
plt.title("Исчезновение интерференции при больших размерах источника")
plt.show()
```
